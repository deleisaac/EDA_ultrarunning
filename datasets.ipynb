```python
#remove h from athelete performance
df_filtered["Athlete performance"] = df_filtered["Athlete performance"].str.split(' ').str.get(0)
df_filtered.head(1)
```

```python
#drop columns: Athelete Club , Athelete Country, Athlete year of birth, Athlete Age Category
df_filtered =  df_filtered.drop(["Athlete club","Athlete country","Athlete year of birth","Athlete age category"],axis=1)
```

```python
#Difference in speed for the 50k,50mi male to female

df_filtered2.groupby(["race_length","athlete_gender"])["athlete_average_speed"].mean()
```

```python
#what age group are the best in the 50km race (20 + races min)
df_filtered2.query("race_length =='50mi'").groupby("athlete_age")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean",ascending=False).query("count>19")
```

```python
df_filtered2.query("race_length =='50mi'").groupby("athlete_age")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean",ascending=True).query("count>19")
```

```python
#seasons for the data > slower in summer than winter?
#spring 3-5
# #summer 6-8
#fall 9-12
#winter 12-2

#splitt between two decimals
df_filtered2["race_month"] = df_filtered2["race_day"].str.split(".").str.get(1).astype(int)
df_filtered2
```

```python
df_filtered2["race_season"] = df_filtered2["race_month"].apply(lambda x: "Winter" if x>11 else "Fall" if x>8 else "Summer" if x>5 else "Spring" if x>2 else "Winter")
```
